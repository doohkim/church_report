```python
import requests
URL = 'https://map.crossfit.com/getAllAffiliates.php'
response = requests.get(URL)
response.text
```

```python
json_type_convert = response.json()
crossfit_affiliate = list()
for i in json_type_convert:
    b=str(i[3])
    crossfit_affiliate.append(b)
len(crossfit_affiliate)
```

# 크롤링

```python
crossfit_affiliate_info = list()
for i in crossfit_affiliate[10000:]:
    URL = 'https://map.crossfit.com/getAffiliateInfo.php?aid=' + i
    response = requests.get(URL)
    crossfit_affiliate_info.append(response.text)
    print(response.text)
```
