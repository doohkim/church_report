```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://korean.visitseoul.net/hallyu?curPage=1')
```

```python
html = res.text
```

```python
header = res.headers
```

```python
header
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
div_paging_lst = soup.find('div', 'paging-lst')
div_paging_lst
```

```python
lastpage_numb = int(div_paging_lst.find_all('a')[-1]['href'].split('=')[-1])
lastpage_numb
```

```python
for num in range(1,lastpage_numb+1):
    url = 'https://korean.visitseoul.net/hallyu?curPage=' + str(num)
#     print(url)
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')
    article_list_slide = soup.find('div', 'article-list-slide')
    ul_article_list = article_list_slide.find('ul', 'article-list')
    for a_tag in ul_article_list.find_all('a'):
        print(a_tag['title'])
        print('https://korean.visitseoul.net/'+a_tag['href'])
        print('\n')

    
```
